```python
import pandas as pd
from nltk.tokenize import TweetTokenizer
fake = pd.read_csv("Downloads/archive/Fake.csv")
real = pd.read_csv("Downloads/archive/True.csv")
fake['label'] = 0
real['label'] = 1
```

```python
true_and_false = pd.concat([fake, real])
```

```python
real_and_fake_text = true_and_false['text']
print(real_and_fake_text)
```

```python
tweet_tokenizer = TweetTokenizer()
tokens = real_and_fake_text.apply(tweet_tokenizer.tokenize)
print(real_and_fake_text, tokens)
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
stopwords_ = set(stopwords.words("english"))
tokens = real_and_fake_text.apply(tweet_tokenizer.tokenize)
clean_tokens = [t for t in set(tokens) if not t in set(stopwords_)]
clean_text = " ".join(clean_tokens)
print(real_and_fake_text, clean_text)
```
